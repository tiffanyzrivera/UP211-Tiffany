# Mkay moving on to spatial visualization of the data!

```python
import geopandas as gpd
```

```python
# read in a geojson file downloaded from the LA Times
tracts=gpd.read_file('data/Census_Tracts_2020.geojson')
tracts.head()
```

```python
# plot it!
tracts.plot(figsize=(12,10))
```

```python
# tell me more about this dataset
tracts.info(verbose=True, show_counts=True)
```

```python
# we only really need FIPS and geometry, so let's subset the data
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
# create a FIPS column

#df['variable']= ''
#if the variable does not already exist, it will create one. If the var already exists it will replace it with what follows the equal sign
tracts['FIPS'] ='06' + '037' + tracts['CT20']
```

```python
# check it!
tracts.head()
```

```python
# create a new dataframe based on the join
#newdf= file_youwantto_keep .merge (df_you_wanna_merge, on= "common variable"
tracts_race=tracts.merge(df2,on="FIPS")
```

I am visualizing where most people of retirement age are located throughout LA using natural breaks. Well, visualizing a subset of the total population of people of retirement age since I could not make a new column with the aggregate values for each census tract. 

```python
tracts_race.plot(figsize=(12,10),
                 column='65 to 74 Years',
                 legend=True, 
                 scheme='NaturalBreaks')
```

Here, I repeat the same process but instead use equal intervals to create the different categories. This map does not show the difference between population as clearly as the one above. 

```python
tracts_race.plot(figsize=(12,10),
                 column='65 to 74 Years',
                 legend=True, 
                 scheme='equal_interval')
#but equal intervals might be the most informative for this type of data
#choose the breakdwosn that make sense for the story you are trying to tell
```

```python
tracts_race.plot(figsize=(12,10),
                 column='65 to 74 Years',
                 legend=True, 
                 scheme='quantiles')

#quantile breaks usaully make the most visually appealing breaks (because they have the same number of census tracts in them)
```

```python
# First, create new columns for percentages
#how to find the proportion of each category
tracts_race['PCT_Elderly'] = tracts_race['Elderly']/tracts_race['Total Population']*100
```

```python
tracts_race.head()
```

```python
tracts_race.plot(figsize=(12,10),
                 column='PCT_Elderly',
                 legend=True, 
                 scheme='equal_interval')
```

```python
tracts_race.plot(figsize=(12,10),
                 column='PCT_Elderly',
                 legend=True, 
                 scheme='NaturalBreaks')
```

Not sure which graph is more informative, but visually the equal interval maps identifies one census tract that has significantly larger proportion of elderly people. 

```python
#where are census tracts where 50% or more are black, but would need to agment this graph to include geolocationtoo
tracts_race[tracts_race.PCT_Elderly > 50].plot(figsize=(12,10))
```

```python
#where are census tracts where 75% or more are black, but would need to agment this graph to include geolocationtoo
tracts_race[tracts_race.PCT_Elderly > 75].plot(figsize=(12,10))
```

## Mkay, lets try to make an interactive map now using folium

```python
import folium
```

```python
m = folium.Map(location=[34.2,-118.2], 
               zoom_start = 9,
               tiles='CartoDB positron', 
               attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
                  geo_data=tracts_race, # geo data
                  data=tracts_race, # data          
                  key_on='feature.properties.FIPS', # key, or merge column
                  columns=['FIPS', 'PCT_Elderly'], # [key, value]
                  fill_color='BuPu',
                  line_weight=0.1, 
                  fill_opacity=0.8,
                  line_opacity=0.2, # line opacity (of the border)
                  legend_name='Population Black (2020)').add_to(m)    # name on the legend color bar
```

```python
m.save('LA_Elderly.html')
```
